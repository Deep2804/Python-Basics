```python
from matplotlib.patches import Ellipse

from math import *

def ellipse(a,b,ra,rb):
    ra=2*ra
    rb=2*rb
    plt.figure()
    ax=plt.gca()
    ellipse=Ellipse(xy=(a,b),width=ra,height=rb)
    ax.add_patch(ellipse)
    plt.axis('scaled')
    plt.show()
```
